Line plots
They are generally used to visualizing the trends over the sequential data  making them particularly useful for time series analysis

```python
# Generate date range

date_range = pd.date_range(start='2024-01-01', end= '2024-12-31', freq='D')

# generate temparatture data (synthetic data)
temperature = np.random.normal(loc=20,scale=5, size=len(date_range))

# Create data frame
time_series_data = pd.DataFrame({'Date': date_range, "Temparature": temperature})

time_series_data.head()
```

```python
# Lineplot with matplotlib

plt.figure(figsize=(12,4))
plt.plot(time_series_data['Date'], time_series_data['Temparature'], marker='', color='b', linewidth=2, linestyle='-')
plt.title("Daily temperature over year")
plt.xlabel("Date")
plt.ylabel("Temperature(C")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Line plot with seaborn
plt.figure(figsize=(12,4))
sns.lineplot(x='Date', y='Temparature', data=time_series_data, marker='o', color='g', linewidth=2, linestyle='-')
plt.title("Daily temperature over year")
plt.xlabel("Date")
plt.ylabel("Temperature(C)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Adding markers and annotations with matplot lib

plt.figure(figsize=(12,4))
plt.plot(time_series_data['Date'], time_series_data['Temparature'], marker='', color='b', linewidth=2, linestyle='-')
plt.title("Daily temperature over year")
plt.xlabel("Date")
plt.ylabel("Temperature(C")
plt.xticks(rotation=45)
plt.grid(True)

# Annotate a specific point

max_temp_date = time_series_data.loc[time_series_data['Temparature'].idxmax(), 'Date']
max_temp = time_series_data['Temparature'].max()
plt.annotate(f'Highest temperature : {max_temp:.2f}C', xy=(max_temp_date, max_temp),xytext=(max_temp_date,max_temp+5),
            arrowprops=dict(facecolor='black',shrink=0.05))
plt.show()
```

```python
# Generate additional synthetic data  for comparision(e.g daily humidity measurements)
humidity = np.random.normal(loc=60, scale=100, size=len(date_range))
time_series_data['Humidity']= humidity
# Multiple Line plot with seaborn
plt.figure(figsize=(12,4))
sns.lineplot(x='Date', y='Temparature', data=time_series_data, marker='o', color='g', linewidth=1, linestyle='-', label='Temperature')
sns.lineplot(x='Date', y='Humidity', data=time_series_data, marker='x', color='b', linewidth=2, linestyle='-', label='Humidity')
plt.title("Daily temperature and humidity over year")
plt.xlabel("Date")
plt.ylabel("Measurements")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Calcualting moving average
time_series_data['Temperature_MA']= time_series_data['Temparature'].rolling(window=7).mean() 

# Line plot with moving average
plt.figure(figsize=(12,4))
plt.plot(time_series_data['Date'], time_series_data['Temparature'], marker='o', color='b', linewidth=1, linestyle='-', alpha=0.5, label='Temperature')
plt.plot(time_series_data['Date'], time_series_data['Temperature_MA'], marker='', color='r', linewidth=2, linestyle='-', label='7-DAY MOVING AVERAGE')
plt.title("Daily temperature over year with moving average")
plt.xlabel("Date")
plt.ylabel("Temperature(C")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
