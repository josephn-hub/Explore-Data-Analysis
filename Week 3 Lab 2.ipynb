# Feature Scaling Technique - StandardScaler

```python
# Loding the library
import pandas as pd
```

```python
# Loading the dataset
heart_df = pd.read_csv("heart.csv")
```

```python
# Printing the first 5 records in dataset
heart_df.head()
```

```python
# checking the dtype
heart_df.dtypes
```

```python
# Taking all the input column in a separate dataframe
```

```python
X = heart_df.drop(columns = ['target'])
```

```python
X
```

```python
# loading the StandardScaler Class
from sklearn.preprocessing import StandardScaler
```

```python
# creating the object of StandardScaler
sc = StandardScaler()
```

```python
# applying the fit() function
sc.fit(X)
```

```python
# applying the transform function
X_new = sc.transform(X)
```

```python
# printing the transformed dataset
pd.DataFrame(X_new)
```
