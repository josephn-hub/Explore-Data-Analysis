```python
# Loading the library
import pandas as pd
```

```python
# Loading the dataset
tips_df = pd.read_csv("tips.csv")
```

```python
tips_df.head()
```

```python
# filtering df where total_bill greater than 30
df = tips_df[tips_df['total_bill'] >30]
```

```python
df
```

# groupby() functionality

```python
# Applying the groupby() fuction on the column - time

G1 = tips_df.groupby('time')
```

```python
# Printing the first entries of each group
G1.first()
```

```python
# Let's find the values present in 'Dinner' group

G1.get_group('Dinner')
```

```python
# The values present in 'Lunch' group
G1.get_group('Lunch')
```

# Aggregate funtions - sum, min, avg etc

```python
# Let's apply 'sum' and 'min' aggregate function.
# 'sum' will give the sum of the individual column and
# 'min' will give the minimum values present in each columns
```

```python
tips_df[['total_bill','tip','size']].aggregate(['sum','min'])
```

```python
# Note : Above operation will only work for the numerical columns
```

```python
# applying the 'average' operation on the column 'total_bill'

tips_df['total_bill'].aggregate(['average'])
```

```python
# applying the 'max' operation on the column 'tip'

tips_df['tip'].aggregate(['max'])
```

# Time Series Operation

```python
# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Create a time series DataFrame with dates over five days
dates = pd.date_range(start='2024-08-01', periods=5, freq='D')
data = [5, 8, 6, 9, 7]
df = pd.DataFrame(data, columns=['Value'], index=dates)

# Plot the data
df.plot(title="Simple Time Series Plot", grid=True)
plt.show()


```
