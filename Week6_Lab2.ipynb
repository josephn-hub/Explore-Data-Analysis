# Plotting charts with seaborn library

```python
# Loading the library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the built-in tips dataset
tips = sns.load_dataset('tips')
tips.info()
```

```python
tips.head()

```

## 1. Count Plot

```python
plt.figure(figsize=(6,4))
sns.countplot(x='day', data=tips)
plt.title('Count Plot')
plt.show()
```

## 2. pairplot

```python
sns.pairplot(tips, hue='sex')
plt.suptitle('Pairplot', y=1.02)
plt.show()
```

## 3. Histplot

```python
plt.figure(figsize=(6,4))
sns.histplot(tips['total_bill'], kde=True, bins=30)
plt.title('Histplot')
plt.show()
```

## 4. Box plot

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot')
plt.show()
```

## 5. Heatmap

```python
plt.figure(figsize=(6,4))
correlation_matrix = tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
# The above error is coming up since there are string values in columns 'sex', 'smoker', 'day', 'time' as we can't do on the correlational matrix
```

```python

#Select only the numeric columns
numeric_tips = tips[['total_bill', 'tip', 'size']]

# PLotting the correlational matrix
plt.figure(figsize=(6,4))
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
# Since there are columns which having string values the correlation matrix showing error.
# Will convert categorical values using get_dummies method
```

```python
# One-hot encode the categorical variables
tips_encoded = pd.get_dummies(tips, columns=['sex', 'smoker', 'day', 'time'])
```

```python
tips_encoded.head()
```

```python
plt.figure(figsize=(12,8))
correlation_matrix = tips_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
